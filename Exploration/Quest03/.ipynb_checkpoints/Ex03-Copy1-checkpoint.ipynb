```python
# 필요한 패키지 import 하기
import os # 환경 변수나 디렉터리, 파일 등의 OS 자원을 제어할 수 있게 해주는 모듈
import cv2 # OpenCV라이브러리 → 컴퓨터 비전 관련 프로그래밍을 쉽게 할 수 있도록 도와주는 라이브러리
import matplotlib.pyplot as plt # 다양한 데이터를 많은 방법으로 도식화 할 수 있도록 하는 라이브러리
import numpy as np # 다차원 배열을 쉽게 처리하고 효율적으로 사용할 수 있도록 하는 라이브러리
import dlib # 이미지 처리 및 기계 학습, 얼굴인식 등을 할 수 있는 c++ 로 개발된 고성능의 라이브러리
print("🌫🛸")
```

```python
#얼굴이미지 rgb 변환
home_dir = os.getenv('HOME')
my_image_path = os.path.join(home_dir, 'work/camera_sticker/images/image.png') # 본인 이미지가 있는 경로를 가지고 온다.
img_bgr = cv2.imread(my_image_path)    # OpenCV로 이미지를 불러옵니다
img_show = img_bgr.copy()      # 출력용 이미지를 따로 보관합니다
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb) # 이미지를 출력하기 위해 출력할 이미지를 올려준다. (실제 출력은 하지 않음)
plt.show() # 이미지를 출력해준다. (실제 출력)
```

```python
# detector를 선언합니다
detector_hog = dlib.get_frontal_face_detector() # 기본 얼굴 감지기를 반환
print("🌫🛸")
```

```python
# bounding box를 추출
#img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
dlib_rects = detector_hog(img_rgb, 1)   # (image, num of image pyramid)
print("🌫🛸")
```

```python
# 찾은 얼굴 영역 박스 리스트
# 여러 얼굴이 있을 수 있습니다
print(dlib_rects)

for dlib_rect in dlib_rects: # 찾은 얼굴 영역의 좌표
    l = dlib_rect.left() # 왼쪽
    t = dlib_rect.top() # 위쪽
    r = dlib_rect.right() # 오른쪽
    b = dlib_rect.bottom() # 아래쪽

    cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA) # 시작점의 좌표와 종료점 좌표로 직각 사각형을 그림

img_show_rgb =  cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

```python
#Dlib의 제공되는 모델 weight file 다운로드 
!wget http://dlib.net/files/shape_predictor_68_face_landmarks.dat.bz2
!mv shape_predictor_68_face_landmarks.dat.bz2 ~/work/camera_sticker/models
!cd ~/work/camera_sticker && bzip2 -d ./models/shape_predictor_68_face_landmarks.dat.bz2
```

```python
#저장한 landmark 모델을 불러오기
model_path = os.path.join(home_dir, 'work/camera_sticker/models/shape_predictor_68_face_landmarks.dat')
    # 저장한 landmark 모델의 주소를 model_path 변수에 저장
landmark_predictor = dlib.shape_predictor(model_path)
    # dlib 라이브러리의 shape_predictor 함수를 이용하여 모델을 불러옴
    # landmark_predictor는 RGB이미지와 dlib.rectangle를 입력 받고 dlib.full_object_detection를 반환
    # dlib.rectangle: 내부를 예측하는 박스
    # dlib.full_object_detection: 각 구성 요소의 위치와, 이미지 상의 객체의 위치를 나타냄
print("🌫🛸")
```

```python
#랜드마크 위치 저장
list_landmarks = []
    # 랜드마크의 위치를 저장할 list 생성

# 얼굴 영역 박스 마다 face landmark를 찾아냅니다
# face landmark 좌표를 저장해둡니다
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)
        # 모든 landmark의 위치정보를 points 변수에 저장
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
        # 각각의 landmark 위치정보를 (x,y) 형태로 변환하여 list_points 리스트로 저장
    list_landmarks.append(list_points)
        # list_landmarks에 랜드마크 리스트를 저장

print(len(list_landmarks[0]))
    # 얼굴이 n개인 경우 list_landmarks는 n개의 원소를 갖고
    # 각 원소는 68개의 랜드마크 위치가 나열된 list
    # list_landmarks의 원소가 1개이므로 list_landmarks[1]을 호출하면 IndexError가 발생
```

```python
#랜드마크 출력
# list_landmarks의 원소가 1개이므로 아래 반복문은 한번만 실행됨
for landmark in list_landmarks:
    for point in landmark:
        cv2.circle(img_show, point, 2, (0, 255, 255), -1)
            # cv2.circle: OpenCV의 원을 그리는 함수
            # img_show 이미지 위 각각의 point에
            # 크기가 2이고 (0, 255, 255)색으로 내부가 채워진(-1) 원을 그림
            # (마지막 인수가 자연수라면 그만큼의 두께의 선으로 원이 그려짐)

img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
    # RGB 이미지로 전환
plt.imshow(img_show_rgb)
    # 이미지를 준비
plt.show()
    # 이미지를 출력
```

```python
#스티커를 그릴 위치 기준값(x,y), 얼굴영역 박스 크기(w,h) 구하기 

for dlib_rect, landmark in zip(dlib_rects, list_landmarks): 
    # 얼굴 영역을 저장하고 있는 값과 68개의 랜드마크를 저장하고 있는 값으로 반복문 실행
    
    print (landmark[30]) # 코의 index는 30 입니다 #437, 182
    x = landmark[30][0] # 이미지에서 코 부위의 x값
    y = landmark[30][1] #예제는 이미시작점이지만 코끝으로 - dlib_rect.height()//2 #이마시작점
 
    
    w = h = dlib_rect.width() 
    # 얼굴 영역의 가로를 차지하는 픽셀의 수  max(x) - min(x) +1
    #(픽셀의 수 이기 때문에 1을 더해줌 → 픽셀 수는 점 하나로도 1이 됨)
    
    print (f'(x,y) : ({x},{y})')
    print (f'(w,h) : ({w},{h})')
```

```python
#고양이 수염 스티커 불러오기 및 크기 지정
sticker_path = os.path.join(home_dir, 'work/camera_sticker/images/cat-whiskers.png') # 수염 이미지의 경로
img_sticker = cv2.imread(sticker_path, cv2.IMREAD_UNCHANGED) 
# 스티커 이미지를 불러옵니다 //
#cv2.imread(이미지 경로) → image객체 행렬을 반환
#img_sticker의 차원확인을 위해서 cv2.IMRED_UNCHANGED

new_width = int(w + w * 0.3)
new_height = int(h + h * 0.3)
img_sticker = cv2.resize(img_sticker, (new_width, new_height))
#img_sticker = cv2.resize(img_sticker, (int(w+w*0.3),int(h+h*0.3))) 
# 스티커 이미지 조정 → w,h는 얼굴 영역의 가로를 차지하는 픽셀의 수
#// cv2.resize(image객체 행렬, (가로 길이, 세로 길이))
#고양이 수염은 얼굴가로사이즈 보다 크면 이쁘기 때문에 (30% 확대)

new_width = int(w * 0.7)
new_height = int(h * 0.7)
img_sticker = cv2.resize(img_sticker, (new_width, new_height))



print (img_sticker.shape) # 사이즈를 조정한 수염 이미지의 차원 확인
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red') # 주 그리드
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black') # 보조 그리드 (더 촘촘)
#수염의 정확한 위치 측정을 위해 그리드 그리기

#plt.grid(True, linestyle='-', linewidth='0.7', color='gray')
plt.imshow(img_sticker)
```

```python
#스티커가 그려질 위치 지정
refined_x = int(x - (w*0.7) // 2)
refined_y = y - 50 #코끝과 스티커이미지의 일치 시킬 위치를 빼줌
# 원본 이미지에 스티커 이미지를 추가하기 위해서 x, y 좌표를 조정합니다. 이미지 시작점은 top-left 좌표이기 때문입니다.
# 즉, refined_x, refined_y값에서 왕관 이미지가 시작됨
print (f'(x,y) : ({refined_x},{refined_y})') # 음수 발생 : 이미지 범위를 벗어남
# 우리는 현재 이마 자리에 왕관을 두고 싶은건데, 이마위치 - 왕관 높이를 했더니 이미지의 범위를 초과하여 음수가 나오는 것
# opencv는 ndarray데이터를 사용하는데, ndarray는 음수인덱스에 접근 불가하므로 스티커 이미지를 잘라 줘야 한다.
```

```python
# 스티커 이미지가 얼굴 이미지 보다 클경우 크롭 처리
if refined_x < 0:
    img_sticker = img_sticker[:, -refined_x:]
    refined_x = 0
# 왕관 이미지를 씌우기 위해 왕관 이미지가 시작할 y좌표 값 조정
if refined_y < 0:
    img_sticker = img_sticker[-refined_y:, :] 
    # refined_y가 -98이므로, img_sticker[98: , :]가 된다. (187, 187, 3)에서 (89, 187, 3)이 됨 (187개 중에서 98개가 잘려나감)
    
    refined_y = 0

print (f'(x,y) : ({refined_x},{refined_y})')
```

```python
# 이미지 스티커 합성, 알파 블랜딩 포함
# 임지 스티커가 4채널 이미지임으로 예제의 3채널 np.whear로 작업할 경우 이미지 오류 발생 
# 따라서 알파 블랜딩 필요 (BGRA)
h_s, w_s = img_sticker.shape[:2]

img_show = img_bgr.copy()#위치 조정시 잔여이미지가 남지 않도록 이미지 다시 불러와서 초기화

# 원본 이미지에서 스티커가 적용될 영역을 잘라서 sticker_area로 저장
sticker_area = img_show[refined_y:refined_y+h_s, refined_x:refined_x+w_s]

if img_sticker.shape[2] == 4:
    # BGR + Alpha 분리
    sticker_bgr = img_sticker[:, :, :3].astype(np.float32)
    sticker_alpha = img_sticker[:, :, 3].astype(np.float32) / 255.0  # shape: (H, W)

    # shape 맞추기 위해 (H, W, 1)로 확장
    sticker_alpha = sticker_alpha[:, :, np.newaxis]

    # 배경 이미지도 float32로 형변환
    sticker_area = sticker_area.astype(np.float32)

    # 알파 블렌딩
    blended = (sticker_bgr * sticker_alpha + sticker_area * (1 - sticker_alpha)).astype(np.uint8)

    # 결과를 원본 이미지에 반영
    img_show[refined_y:refined_y+h_s, refined_x:refined_x+w_s] = blended

else:
    # 예외 처리: 4채널이 아닐 경우 fallback (3채널용)
    mask = (img_sticker == 0).all(axis=2, keepdims=True)
    #스티커 이미지의 모든 채널 값이 0인 픽샐(검은색)을 찾아 마스크 생성
    out = np.where(mask, sticker_area, img_sticker).astype(np.uint8)
    #마스크가 True(검은색)인 부분에는 sticker_area(배경), False는 img_stiker을 합성
    img_show[refined_y:refined_y+h_s, refined_x:refined_x+w_s] = out
    #합성된결과를 img_show에 씌움

# 시각화
plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
