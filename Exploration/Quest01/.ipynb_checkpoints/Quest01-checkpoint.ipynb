```python
#당뇨병 데이터 가져오기
from sklearn.datasets import load_diabetes
data=load_diabetes()
df_X=data.data
df_y=data.target
print(data.feature_names)
```

```python
df_X_pd = pd.DataFrame(df_X, columns=data.feature_names)
print("\n변환된 df_X DataFrame의 상위 5행:")
print(df_X_pd.head())
#특성 수치를 직접 보고 싶었음. 이 수치에 따라 스케일링 작업등을 고려했으나 이미 되어 있다고 판단
```

```python
# 각 특성의 상관 관계 확인

import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes


# 특성 데이터(df_X)를 DataFrame으로 변환
#    컬럼 이름을 붙여서 어떤 특성과 y의 상관관계를 보는지 명확하게 합니다.
df_X_pd = pd.DataFrame(df_X, columns=data.feature_names)

# 타겟 데이터(df_y)를 Series로 변환
df_y_pd = pd.Series(df_y, name='target')

# 특성 데이터와 타겟 데이터를 하나의 DataFrame으로 합치기
#    axis=1은 컬럼 방향으로 합친다는 의미입니다.
df_combined = pd.concat([df_X_pd, df_y_pd], axis=1)

# 상관계수 계산하기
#    .corr() 메서드는 DataFrame 내의 모든 컬럼 쌍에 대한 상관계수를 계산합니다.
correlation_matrix = df_combined.corr()

# 각 특성과 'target' 간의 상관계수만 추출하여 출력
#    'target' 컬럼과 다른 모든 컬럼들의 상관계수를 보여줍니다.
print("각 특성과 타겟(y) 간의 상관계수:\n")
print(correlation_matrix['target'].sort_values(ascending=False))
# 상관계수가 낮은 sex,S2,S1,age 등을 특성에서 제거해보았으나 결과적으로 차이 없음
# 제거 작업은 비교를 위해 코드 마지막에 남겨두었음
```

```python
# 데이터 총 수량 확인
print(df_X.shape)
print(df_y.shape)
```

```python
#모델에 입력할 X 준비하기
import numpy as np

X=np.array(df_X)

print(type(X), X.dtype)
```

```python
#모델에 예측할 데이터 y 준비하기
y=np.array(df_y)

print(type(y), y.dtype)
```

```python
#train 데이터와 test 데이터로 분리하기
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=21
)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
#모델 준비하기 
W = np.random.rand(X_train.shape[1])
b = np.random.rand()

print(len(W))
print(b)
```

```python
#모델 불러오기

def model(X, W, b):
    predictions = 0
    for i in range(len(W)):
        predictions += X[:, i] * W[i]
    predictions += b
    return predictions
```

```python
#손실함수 mse 
def MSE(a, b):
    mse = ((a - b) ** 2).mean()  # 두 값의 차이의 제곱의 평균
    return mse

def loss(X, W, b, y):
    predictions = model(X, W, b)
    L = MSE(predictions, y)
    return L
```

```python
#기울기를 구하는 gradient 함수 구현하기
def gradient(X, W, b, y):
    N = len(y)

    y_pred = model(X, W, b)

    dW = 1/N * 2 * X.T.dot(y_pred - y)

    db = 2 * (y_pred - y).mean()
    return dW, db
```

```python
#하이퍼 파라미터 학습률 설정하기 
LEARNING_RATE = 0.3 # 0.001 에서 0.01까지 수정하면서 실험
```

```python
#모델 학습하기
global W, b

losses = []

for i in range(1, 100001): # 2,000에서 3,000/5,000/10,000/50,000 으로 epoch 높이면서 실험
    dW, db = gradient(X_train, W, b, y_train)
    W -= LEARNING_RATE * dW
    b -= LEARNING_RATE * db
    L = loss(X_train, W, b, y_train)
    losses.append(L)
    if i % 100 == 0:
        print('Iteration %d : Loss %0.4f' % (i, L))
```

```python
#test데이터에 대한 성능 확인
prediction = model(X_test, W, b)
mse = loss(X_test, W, b, y_test)
mse
```

```python
#정답 데이터와 예측한 데이터 시각화 하기
import matplotlib.pyplot as plt

plt.scatter(X_test[:, 0], y_test)
plt.scatter(X_test[:, 0], prediction)
plt.show()
```

```python
# 이후 특성에서 상관계수가 낮은 특성을 제거하고 모델 학습을 시키는 실험을 하였음
 # 결과적으로 큰 차이 없음
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
import pandas as pd # 확인용


# numpy 배열 X, y 준비 (기존 코드)
X = np.array(df_X)
y = np.array(df_y)

X_train_full, X_test_full, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=21
)

print(f"원본 X_train.shape: {X_train_full.shape}") # (353, 10)

# 2. 제거할 컬럼의 인덱스를 지정
# 'sex' 컬럼 (인덱스 1)을 제거할 예정
columns_to_remove_indices = [4] # 리스트 형태로 지정

# 3. train 및 test 데이터셋에서 해당 컬럼 제거
# np.delete 함수 사용: axis=1은 컬럼을 의미
X_train = np.delete(X_train_full, columns_to_remove_indices, axis=1)
X_test = np.delete(X_test_full, columns_to_remove_indices, axis=1)

print(f"특성 제거 후 X_train.shape: {X_train.shape}") # (353, 9) (sex 제거)
print(f"특성 제거 후 X_test.shape: {X_test.shape}")   # (89, 9)

# 4. 모델 파라미터 (W) 재초기화
# 특성 수가 10개에서 9개로 줄었으므로 W의 크기도 맞춰야 합니다.
W = np.random.rand(X_train.shape[1]) # 변경된 특성 수(9)에 맞게 W 초기화
b = np.random.rand()

print(f"새로운 W의 길이: {len(W)}") # 9

# 이어서 학습 코드를 실행하면 됩니다.
# global W, b (W와 b 초기화가 같은 셀에서 이루어지면 필요없지만, 다른 셀이라면 유지)
# ... 학습 루프 코드 ...
```

```python
#손실함수 mse 
def MSE(a, b):
    mse = ((a - b) ** 2).mean()  # 두 값의 차이의 제곱의 평균
    return mse
```

```python
def loss(X, W, b, y):
    predictions = model(X, W, b)
    L = MSE(predictions, y)
    return L
```

```python
#기울기를 구하는 gradient 함수 구현하기
def gradient(X, W, b, y):
    N = len(y)

    y_pred = model(X, W, b)

    dW = 1/N * 2 * X.T.dot(y_pred - y)

    db = 2 * (y_pred - y).mean()
    return dW, db
```

```python
#하이퍼 파라미터 학습률 설정하기 
LEARNING_RATE = 0.3 # 0.001 에서 0.01까지 수정하면서 실험

#모델 학습하기
global W, b

losses = []

for i in range(1, 100001): # 1,000에서 3,000/5,000/10,000/50,000 으로 epoch 높이면서 실험
    dW, db = gradient(X_train, W, b, y_train)
    W -= LEARNING_RATE * dW
    b -= LEARNING_RATE * db
    L = loss(X_train, W, b, y_train)
    losses.append(L)
    if i % 100 == 0:
        print('Iteration %d : Loss %0.4f' % (i, L))
```

```python
#데이터 가져오기
import pandas as pd

train = pd.read_csv("~/data/data/bike-sharing-demand/train.csv") 
train.info()
```

```python
#datdtime 컬럼을 datdtime 자료형으로 변환
#연,월,일,시,분,초까지 6가지 컬럼 생성
train['datetime'] = pd.to_datetime(train['datetime'])
train['year'] = train['datetime'].dt.year
train['month'] = train['datetime'].dt.month
train['day'] = train['datetime'].dt.day
train['hour'] = train['datetime'].dt.hour
train['minute'] = train['datetime'].dt.minute
train['second'] = train['datetime'].dt.second

train.info()
train.head()
```

```python
# 각 season이 포함하는 월을 체크 
season_month_mapping = train.groupby('season')['month'].unique().apply(list)
print("\n각 Season에 포함된 월(month):")
print(season_month_mapping)
```

```python
# 월별 온도 체크
# 겨울시즌 월 구성이 이상하다 [1,2,3] 12월을 겨울이고 3월은 봄으로 봐야할 듯하다.
# 지역별 특성이 있을 수 있으니 월별 평균 온도를 확인해본다
# month를 기준으로 그룹화하고 'temp'의 평균을 계산.
monthly_avg_temp = train.groupby('month')['temp'].mean()

print("각 월별 평균 온도:")
print(monthly_avg_temp)
```

```python
# season 재구성하기 
 # 월별 온도 역시 12,1,2월이 가장 낮다.
 # 기존 1(겨울) [1,2,3] 을 [12,1,2] 등으로 변경

conditions=[
    (train['month']==12) | (train['month']==1) | (train['month']==2),
    (train['month']>=3) & (train['month']<=5),
    (train['month']>=6) & (train['month']<=8),
    (train['month']>=9) & (train['month']<=11)
]
     
choices=[1,2,3,4]
# numpy.select를 사용하여 새로운 'season_redefined' 컬럼 생성
train['season_redefined'] = np.select(conditions, choices, default=np.nan) # default는 모든 조건에 해당하지 않을 경우의 값

print("새로운 'season_redefined' 컬럼이 추가된 train.head():")
print(train[['month', 'season', 'season_redefined']].head()) # 기존 season 컬럼도 함께 확인

# 새로운 시즌 매핑 확인
new_season_month_mapping = train.groupby('season_redefined')['month'].unique().apply(list)
print("\n새로운 Season에 포함된 월(month):")
print(new_season_month_mapping)
```

```python
# year,month 등 datetime관련 요소 데이터 개수 시각화
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(18,6))
sns.boxplot(x= 'month', y='count', data=train)

plt.figure(figsize=(18,6))
sns.boxplot(x= 'temp', y='count', data=train)
plt.xticks(rotation=90)

plt.figure(figsize=(18,6))
sns.boxplot(x= 'humidity', y='count', data=train)
plt.xticks(rotation=90)

plt.figure(figsize=(18,6))
sns.boxplot(x= 'windspeed', y='count', data=train)
plt.xticks(rotation=90)

plt.figure(figsize=(18,6))
sns.boxplot(x= 'holiday', y='count', data=train)

plt.figure(figsize=(18,6))
sns.boxplot(x= 'hour', y='count', data=train)

plt.figure(figsize=(18,6))
sns.boxplot(x= 'season', y='count', data=train)


fig, axs = plt.subplots(8, 1, figsize=(18, 50))
cols = ['month', 'holiday', 'season', 'year', 'day', 'hour', 'minute', 'second']
for ax, col in zip(axs.flatten(), cols):
    sns.countplot(x=col, data=train, ax=ax)
    ax.set_title(f'Countplot of {col}')
plt.tight_layout()
plt.show()





```

```python
# X, y 컬럼 선택 및 train/test 데이터 분리
features = ['year', 'month', 'day', 'hour', 'temp', 'humidity', 'windspeed', 'holiday', 'season_redefined']
bike_X = train[features]
bike_y = train['count']

from sklearn.model_selection import train_test_split
bike_X_train, bike_X_test, bike_y_train, bike_y_test = train_test_split(
    bike_X, bike_y, test_size=0.2, random_state=42
)
print(bike_X_train.shape)
print(bike_X_test.shape)
print(bike_y_train.shape)
print(bike_y_test.shape)
```

```python
# LinearRegression 모델 불러오기 및 학습
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(bike_X_train, bike_y_train)
```

```python
# bike_X_test 예측 값 및 손실함수 MSE & RMSE 계산
from sklearn.metrics import mean_squared_error
import numpy as np

bike_y_pred = model.predict(bike_X_test)
mse = mean_squared_error(bike_y_test, bike_y_pred)
rmse = np.sqrt(mse)
print("MSE:", mse)
print("RMSE:", rmse)
```

```python
# X축은 temp or humidity로 y축은 count로 예측 결과 시각화
fig, axs = plt.subplots(2, 2, figsize=(16, 12))

# temp vs count 시각화
axs[0,0].scatter(bike_X_test['temp'], bike_y_test, color='blue', label='Actual')
axs[0,0].scatter(bike_X_test['temp'], bike_y_pred, color='red', label='Predicted', alpha=0.5)
axs[0,0].set_xlabel('temp')
axs[0,0].set_ylabel('count')
axs[0,0].set_title('Temperature vs Count')
axs[0,0].legend()

# humidity vs count 시각화
axs[0,1].scatter(bike_X_test['humidity'], bike_y_test, color='blue', label='Actual')
axs[0,1].scatter(bike_X_test['humidity'], bike_y_pred, color='red', label='Predicted', alpha=0.5)
axs[0,1].set_xlabel('humidity')
axs[0,1].set_ylabel('count')
axs[0,1].set_title('Humidity vs Count')
axs[0,1].legend()

# season_redefined vs count 시각화
axs[1,0].scatter(bike_X_test['season_redefined'], bike_y_test, color='blue', label='Actual')
axs[1,0].scatter(bike_X_test['season_redefined'], bike_y_pred, color='red', label='Predicted', alpha=0.5)
axs[1,0].set_xlabel('season_redefined')
axs[1,0].set_ylabel('count')
axs[1,0].set_title('season_redefined vs Count')
axs[1,0].legend()

# hour vs count 시각화
axs[1,1].scatter(bike_X_test['hour'], bike_y_test, color='blue', label='Actual')
axs[1,1].scatter(bike_X_test['hour'], bike_y_pred, color='red', label='Predicted', alpha=0.5)
axs[1,1].set_xlabel('hour')
axs[1,1].set_ylabel('count')
axs[1,1].set_title('hour vs Count')
axs[1,1].legend()
```
